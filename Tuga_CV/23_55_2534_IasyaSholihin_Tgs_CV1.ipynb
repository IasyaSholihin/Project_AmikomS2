```python

# Baca gambar segmentasi warna untuk mengekstraksi fitur-fitur
image_path = '/content/drive/MyDrive/Image_proc/saya1.jpg'
image = cv2.imread(image_path)

# Konversi gambar ke ruang warna HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Tentukan batas-batas untuk rentang warna kucing (misalnya, warna coklat)
lower_bound = np.array([0, 50, 50])  # Rentang warna bawah untuk coklat dalam HSV
upper_bound = np.array([30, 255, 255])  # Rentang warna atas untuk coklat dalam HSV

# Lakukan segmentasi warna
mask = cv2.inRange(hsv_image, lower_bound, upper_bound)

# Aplikasikan mask pada gambar asli untuk mendapatkan hasil segmentasi
segmented_image = cv2.bitwise_and(image, image, mask=mask)

# Tampilkan gambar asli dan hasil segmentasi
cv2_imshow(image)
cv2_imshow(segmented_image)
```

```python
# Baca gambar Deteksi tepi menggunakan metode Canny ( edge detection )
image_path = '/content/drive/MyDrive/Image_proc/saya1.jpg'   #
image = cv2.imread(image_path)

# Ubah gambar menjadi citra grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Deteksi tepi menggunakan metode Canny
edges = cv2.Canny(gray_image, 100, 90)  # Anda dapat menyesuaikan nilai ambang sesuai kebutuhan

# Tampilkan gambar asli
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Tampilkan hasil deteksi tepi
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection (Canny)')
plt.axis('off')

plt.show()
```
